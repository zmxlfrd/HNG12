### Observations <a id="observations"></a>

1. Top-Selling products <a id="top-selling-products"><a/>

```python
top_selling_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

top_selling_products.head(5)
```

```python
sns.barplot(x=top_selling_products.values, y=top_selling_products.index, palette="Blues_d")
plt.title("Top-Selling Products by Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.show()
```

<d>

- The highest revenue-generating products are `Phones` with \$330,007, followed by `Chairs` with \$328,449 and Storage with \$223,843.
- These categories should be prioritized for marketing campaigns.

<d>

2. Most profitable products <a id="most-profitable-products"><a/>

```python
most_profitable_products = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
most_profitable_products.head()
```

```python
sns.barplot(x=most_profitable_products.values, y=most_profitable_products.index, palette="Greens_d")
plt.title('Most Profitable Products by Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('Sub-Category')
plt.show()
```

- Copiers (55,617) lead in profit despite lower total sales, suggesting high margins.
- Other profitable products include Phones (44,515), Accessories (41,936), and Paper (34,053), indicating their value in sustaining overall business profitability.

<d>

3. Sales Distribution by Region <a id="sales-by-region"><a/>

```python
sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
sales_by_region.head()
```

```python
sns.barplot(x=sales_by_region.values, y=sales_by_region.index, palette="coolwarm")
plt.title('Sales Distribution Across Regions')
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.show()
```

- The West region (725,457) records the highest sales, followed by East (678,781), while the South (391,721) has the lowest.
- Expanding marketing efforts in the South region could boost revenue.
